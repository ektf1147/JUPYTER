```python
import os
os.environ['PROJ_LIB']='/home/jhlee/anaconda3/envs/JUPYTER/share/proj'
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt
import numpy as np

# plot Taal volcano
plt.figure(figsize=(10,10))
llcrnrlat = 12
urcrnrlat = 21
llcrnrlon = 118
urcrnrlon = 125

map = Basemap(projection='cyl', llcrnrlat= llcrnrlat, urcrnrlat = urcrnrlat,\
              llcrnrlon = llcrnrlon, urcrnrlon = urcrnrlon, resolution = 'h')
map.drawcoastlines()
map.fillcontinents(color='white', lake_color='aqua')

map.drawparallels(np.arange(llcrnrlat,urcrnrlat+1,1.), labels=[True,False,False,False])
map.drawmeridians(np.arange(llcrnrlon,urcrnrlon+1,1.), labels=[False,False,False,True])
map.drawmapboundary(fill_color='aqua')

# location of tall volcano
lons = [120.57, 120.40, 120.32, 120.45, 120.57, 121.13, 121.06]
lats = [13.56, 14.02, 14.26, 15.04, 15.07, 14.29, 14.01]
map.drawgreatcircle(lons[0], lats[0], lons[1], lats[1], linewidth=2, color='red')
map.drawgreatcircle(lons[1], lats[1], lons[2], lats[2], linewidth=2, color='red')
map.drawgreatcircle(lons[2], lats[2], lons[3], lats[3], linewidth=2, color='red')
map.drawgreatcircle(lons[3], lats[3], lons[4], lats[4], linewidth=2, color='red')
map.drawgreatcircle(lons[4], lats[4], lons[5], lats[5], linewidth=2, color='red')
map.drawgreatcircle(lons[5], lats[5], lons[6], lats[6], linewidth=2, color='red')
map.drawgreatcircle(lons[6], lats[6], lons[0], lats[0], linewidth=2, color='red')
x, y = map(lons, lats)
#map.plot(x, y, 'ro', markersize=3)


# show result plot
plt.title('Taal volcano')
plt.show()
```
