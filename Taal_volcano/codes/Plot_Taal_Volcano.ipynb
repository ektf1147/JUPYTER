```python
import os
os.environ['PROJ_LIB']='/home/jhlee/anaconda3/envs/JUPYTER/share/proj'
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt
import numpy as np

# plot Taal volcano
plt.figure(figsize=(8,8))
llcrnrlat = 10
urcrnrlat = 30
llcrnrlon = 115
urcrnrlon = 135

map = Basemap(projection='cyl', llcrnrlat= llcrnrlat, urcrnrlat = urcrnrlat,\
              llcrnrlon = llcrnrlon, urcrnrlon = urcrnrlon, resolution = 'h')
map.drawcoastlines()
map.fillcontinents(color='white', lake_color='aqua')

map.drawparallels(np.arange(llcrnrlat,urcrnrlat+1,5.), labels=[True,False,False,False])
map.drawmeridians(np.arange(llcrnrlon,urcrnrlon+1,5.), labels=[False,False,False,True])
map.drawmapboundary(fill_color='aqua')

# show result plot
plt.title('Taal volcano')
plt.show()
```
