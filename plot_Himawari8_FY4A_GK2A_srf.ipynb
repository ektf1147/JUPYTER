```python
import numpy as np
import matplotlib.pyplot as plt


# Read FY-4A SRF
dir = '/storage1/jhlee/NMSC_2018/'
fname = dir + 'FY_4A_AGRI/FY4A_AGRI_065_SRF.txt'
wav_fy4a, srf_fy4a = np.loadtxt(fname, unpack=True)
wav_fy4a = wav_fy4a / 1000. # nm to um
srf_fy4a = srf_fy4a / 100.  # srf => 0 to 1


# Read Himawari8 SRF
fname = dir + 'Himawari_8_AHI/Himawari8_AHI_064_SRF.txt'
wav_hima, srf_hima = np.loadtxt(fname, unpack=True)
srf_hima = srf_hima / 100. # srf => 0 to 1


# Read GK-2A SRF
fname = dir + 'GK_2A_AMI/GK2A_AMI_064_SRF.txt'
wav_gk2a, wavnum_gk2a, srf_gk2a = np.loadtxt(fname, unpack=True)


# Plot SRF
plt.plot(wav_fy4a, srf_fy4a, label='FY-4A')
plt.plot(wav_hima, srf_hima, color='red', label='Himawari-8')
plt.plot(wav_gk2a, srf_gk2a, color='black', label='GK-2A')
plt.xlabel('Wavelength [um]')
plt.ylabel('Spectral response function')
plt.ylim(0,1)

plt.legend()
plt.savefig('./result_plots/SRF_FY4A_Himawari8_GK2A.png', dpi=300)
```
